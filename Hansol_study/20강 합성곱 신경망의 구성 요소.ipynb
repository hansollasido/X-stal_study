- 밀집층에 있는 뉴런의 가중치가 모두 다르듯이 합성곱 층에 있는 필터의 가중치도 모두 다름

# 케라스 합성곱 층

```
from tensorflow import keras
keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu')
```

- 맨앞 10은 채널, 필터 갯수

# 패딩과 스트라이드

- 패딩 : 입력 배열의 주위를 가상의 원소로 채우는 것
- 세임 패딩 : 입력 주위에 0으로 패딩 하는 것
- 밸리드 패딩 : 패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만드는 경우

image.png

- padding의 기본값은 valid


```
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding='same')
```

- 이동의 크기를 **스트라이드**라 함

```
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding='same', strides=1)
```

# 풀링
- 풀링 : 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할
- 차원의 개수는 유지함

image.png

- 최대 풀링
- 평균 풀링

```
keras.layers.MaxPooling2D(2, strides=2, padding='valid')
```

- 평균 풀링을 제공하는 클래스는 AveragePooling2D
