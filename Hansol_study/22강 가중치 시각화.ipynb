# 가중치 시각화

- 합성곱 층 : 여러 개의 필터를 사용해 이미지에서 특징을 학습함
- 각 필터 : 커널이라 부르는 가중치와 절편을 가지고 있음

image.png

```python
from tensorflow import keras
model = keras.models.load_model('best-cnn-model.h5')
```

```python
model.layers
```

- 층 확인 가능

```python
conv = model.layers[0]
print(conv.weights[0].shape, conv.weights[1].shape)
```

- 커널 크기 3,3 에 입력의 깊이가 1, 필터의 개수 32개
- weights의 첫 번째 원소인 가중치의 크기는 (3, 3, 1, 32)

```python
conv_weights = conv.weights[0].numpy()
print(conv_weights.mean(), conv_weights.std())
```

- 이 가중치의 평균값은 0, 표준편차는 0.27
- 가중치가 어떤 분포를 가졌는지 직관적으로 이해하기 쉽도록 히스토그램 그려 보겠음

```python
import matplotlib.pyplot as plt
plt.hist(conv_weights.reshape(-1,1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```
