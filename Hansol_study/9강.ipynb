```python
import pandas as pd
fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

```python
fish['Species'].unique()
```

```python
fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()
```

```python
fish_input[:5]
```

```python
# [[]] 괄호 2개 안치도록 주의
fish_target = fish['Species'].to_numpy()
fish_target
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_input,
                                                                      fish_target, random_state = 42)
```

```python
# 표준화 전처리해야함
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
test_scaled
```

```python
train_target
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(train_scaled, train_target)
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

- 2 개 이상의 클래스가 포함된 문제를 **다중분류**라고 부름

- 사이킷런에서는 문자열로 된 타깃값을 그대로 사용가능

```python
print(kn.classes_)
```

```python
print(kn.predict(test_scaled[:5]))
```

```python
import numpy as np
proba = kn.predict_proba(test_scaled[:5])
print(np.round(proba,decimals=4)) 
```

```python
distances, indexes = kn.kneighbors(test_scaled[3:4])
print(train_target[indexes])
```

```python
bream_smelt_indexes = (train_target ==  'Bream') | (train_target == 'Smelt')
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
```

- bream 또는 smelt 인 정보만 취함

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
print(lr.predict(train_bream_smelt[:5]))
```

- 샘플 예측의 예측 확률

```python
print(lr.predict_proba(train_bream_smelt[:5]))
```

```python
print(lr.classes_)
```

```python
print(lr.coef_,lr.intercept_)
```

image.png

- z값 출력 가능

```python
decisions = lr.decision_function(train_bream_smelt[:5])
print(decisions)
```

- 시그모이드 함수를 통하여 z를 확률화 시킴

```python
from scipy.special import expit
print(expit(decisions))
```

- 클래스를 사용해 7 개의 생선을 분류해 보면서 이진 분류와의 차이점을 알아보겠음

- max_iter 매개변수를 지정하여 반복 횟수를 결정함
  - 기본값은 100임
  - 반복 횟수가 부족하다는 경고 발생하므로 1000으로 늘리겠음

- 릿지 회귀와 마찬가지로 계수의 제곱을 규제함
  - L2 규제라고 부름
  - LogisticRegression 규제를 제어하는 매개변수는 C
  - C는 alpha와 반대로 작을수록 규제가 커짐
  - C의 기본값은 1임


```python
lr = LogisticRegression(C=20, max_iter=1000)
lr.fit(train_scaled,train_target)
print(lr.score(train_scaled,train_target))
print(lr.score(test_scaled,test_target))
```

```python
print(lr.predict(test_scaled[:5]))
```

```python
proba = lr.predict_proba(test_scaled[:5])
print(np.round(proba,decimals=3))
```

```python
print(lr.classes_)
```

```python
# class가 7개이기 때문에 7개가 나타남
print(lr.coef_.shape,lr.intercept_.shape)
```

- 이진 분류와 달리 다중 분류는 소프트맥스(softmax) 함수를 이용하여 z 값을 확률로 변환함

image.png

```python
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals=2))
```

```python
from scipy.special import softmax
proba = softmax(decision, axis=1) # axis를 설정하지 않으면 전체에 대한 확률을 계산하게 됨
print(np.round(proba, decimals=3))
```
